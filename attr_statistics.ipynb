```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('Walt_disney_movie_dataset.csv')

df.head()
```

```python
# Statisticals of Running Time

running_time = df["Running time (int)"]
rt_max = running_time.max()
rt_min = running_time.min()
rt_mean = running_time.mean()
rt_median = running_time.median()
rt_mode = running_time.mode()
print(rt_max)
print(rt_min)
print(rt_mean)
print(rt_median)
print(rt_mode.values)
```

```python
# Statisticals of Box Office

box_office = df["Box office (float)"]
bo_max = box_office.max()
bo_min = box_office.min()
bo_mean = box_office.mean()
bo_median = box_office.median()
bo_mode = box_office.mode()

print(bo_max)
print(bo_min)
print(bo_mean)
print(bo_median)
print(bo_mode.values)
```

```python
# Statisticals of Box Office

budget = df["Budget (float)"]
bg_max = budget.max()
bg_min = budget.min()
bg_mean = budget.mean()
bg_median = budget.median()
bg_mode = budget.mode()

print(bg_max)
print(bg_min)
print(bg_mean)
print(bg_median)
print(bg_mode.values)
```
