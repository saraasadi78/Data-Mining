```python
import pandas as pd
import numpy as np
from scipy.stats import skew
import matplotlib.pyplot as plt

df = pd.read_csv('Walt_disney_movie_dataset.csv')
```

```python
running_time = df["Running time (int)"]
rt_skewness = skew(running_time, nan_policy='omit')
rt_skewness_str = "skewness= " + str(rt_skewness)
plt.hist(running_time, bins=100)
plt.xlabel('Running Time')
plt.ylabel('Counts')
plt.text(40, 25, rt_skewness_str)
plt.show()
```

```python
box_office = df["Box office (float)"]
bo_skewness = skew(box_office, nan_policy='omit')
bo_skewness_str = "skewness= " + str(bo_skewness)
plt.hist(box_office, bins=20)
plt.xlabel('Box Office')
plt.ylabel('Counts')
plt.text(500000000, 80, bo_skewness_str)
plt.show()
```

```python
budget = df["Budget (float)"]
bg_skewness = skew(budget, nan_policy='omit')
bg_skewness_str = "skewness= " + str(bg_skewness)
plt.hist(budget, bins=50)
plt.xlabel('Budget')
plt.ylabel('Counts')
plt.text(50000000, 50, bg_skewness_str)
plt.show()
```
