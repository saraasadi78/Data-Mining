```python
import pandas as pd
import numpy as np
from scipy.spatial.distance import pdist, squareform
import matplotlib.pyplot as plt

df = pd.read_csv('Walt_disney_movie_dataset.csv')
```

```python
running_time = df["Running time (int)"]
box_office = df["Box office (float)"]
budget = df["Budget (float)"]
data_subset = pd.concat([running_time, box_office, budget],axis=1)
# data_subset = [budget, box_office, running_time]
type(data_subset)
data_subset
```

```python
dis_vector = pdist(data_subset, metric='matching')
dis_matrix = squareform(dis_vector,force='tomatrix')
print(dis_matrix)
x = dis_matrix.shape
x = x[0]
count =0
for i in range(0,x):
    for j in range(0,x):
        if dis_matrix[i][j]!=0 and dis_matrix[i][j]!=1:
            print("row: ",i," col: ",j, "value: ",dis_matrix[i][j])
            count+=1
print(count/2)
```
