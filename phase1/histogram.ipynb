```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('Walt_disney_movie_dataset.csv')
```

```python
# Running Time histogram

running_time = df["Running time (int)"]
plt.hist(running_time, bins=167)
plt.title("Running Time Histogram")
plt.xlabel('Running Time')
plt.ylabel('Counts')
plt.grid(True)
plt.show()
```

```python
# Box Office histogram

box_Office = df["Box office (float)"]
plt.hist(box_Office, bins=50)
plt.title("Box Office Histogram")
plt.xlabel('Box Office')
plt.ylabel('Counts')
plt.grid(True)
plt.show()
```

```python
# Budget histogram

budget = df["Budget (float)"]
plt.hist(budget, bins=100)
plt.title("Budget Histogram")
plt.xlabel('Budget')
plt.ylabel('Counts')
plt.grid(True)
plt.show()
```
